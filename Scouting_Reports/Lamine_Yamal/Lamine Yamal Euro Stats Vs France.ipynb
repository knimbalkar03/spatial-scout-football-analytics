```python
successful = passes[
    passes['pass_outcome'].isna()
].copy()
```

```python
import numpy as np

successful['distance_to_goal_start'] = np.sqrt(
    (120 - successful['start_x'])**2 +
    (40 - successful['start_y'])**2
)

successful['distance_to_goal_end'] = np.sqrt(
    (120 - successful['end_x'])**2 +
    (40 - successful['end_y'])**2
)

successful['progression'] = (
    successful['distance_to_goal_start']
    - successful['distance_to_goal_end']
)
```

```python
progressive_passes = successful[
    successful['progression'] >= 10
]

print(len(progressive_passes))
```

```python
from mplsoccer import Pitch, VerticalPitch
import matplotlib.pyplot as plt

# =====================================
# DATA PREP
# =====================================

touches = yamal[yamal['location'].notna()].copy()
touches['x'] = touches['location'].apply(lambda x: x[0])
touches['y'] = touches['location'].apply(lambda x: x[1])

shots = yamal[yamal['type'] == 'Shot'].copy()
shots['x'] = shots['location'].apply(lambda x: x[0])
shots['y'] = shots['location'].apply(lambda x: x[1])

passes = yamal[yamal['type'] == 'Pass'].copy()

goal_shots = shots[shots['shot_outcome'] == 'Goal']
other_shots = shots[shots['shot_outcome'] != 'Goal']

# =====================================
# METRICS
# =====================================

ball_receipts = len(yamal[yamal['type'] == 'Ball Receipt*'])

total_passes = len(passes)

completed_passes = len(
    passes[passes['pass_outcome'].isna()]
)

pass_accuracy = round(
    completed_passes / total_passes * 100,
    1
)

progressive_pass_count = len(progressive_passes)
progressive_carry_count = len(progressive_carries)

progressive_pass_pct = round(
    progressive_pass_count / total_passes * 100,
    1
)

total_carries = len(
    yamal[yamal['type'] == 'Carry']
)

progressive_carry_pct = round(
    progressive_carry_count / total_carries * 100,
    1
)

# =====================================
# FIGURE
# =====================================

fig = plt.figure(
    figsize=(16,10),
    facecolor='#071A2E'
)

# =====================================
# HEADER
# =====================================

fig.text(
    0.5,
    0.975,
    "LAMINE YAMAL vs FRANCE",
    ha='center',
    color='white',
    fontsize=24,
    fontweight='bold'
)

fig.text(
    0.5,
    0.945,
    "UEFA Euro 2024 Semi-Final | The Spatial Scout",
    ha='center',
    color='lightgray',
    fontsize=11
)

# =====================================
# HEATMAP
# =====================================

ax1 = fig.add_axes([0.05,0.52,0.42,0.35])

pitch = Pitch(
    pitch_type='statsbomb',
    pitch_color='#071A2E',
    line_color='white',
    linewidth=2
)

bin_statistic = pitch.bin_statistic_positional(
    touches['x'],
    touches['y'],
    statistic='count'
)

pitch.heatmap_positional(
    bin_statistic,
    ax=ax1,
    cmap='Blues'
)

pitch.draw(ax=ax1)

ax1.set_title(
    "Touch Heatmap",
    color='white',
    fontsize=14
)

# =====================================
# PROGRESSIVE PASSES
# =====================================

ax2 = fig.add_axes([0.53,0.52,0.42,0.35])

pitch.draw(ax=ax2)

pitch.scatter(
    progressive_passes['start_x'],
    progressive_passes['start_y'],
    color='white',
    s=40,
    ax=ax2
)

pitch.scatter(
    progressive_passes['end_x'],
    progressive_passes['end_y'],
    color='cyan',
    s=70,
    ax=ax2
)

pitch.arrows(
    progressive_passes['start_x'],
    progressive_passes['start_y'],
    progressive_passes['end_x'],
    progressive_passes['end_y'],
    color='cyan',
    width=2,
    ax=ax2
)

ax2.text(
    0.5,
    -0.08,
    "White = Start | Cyan = End",
    transform=ax2.transAxes,
    ha='center',
    color='white',
    fontsize=8
)

ax2.set_title(
    "Progressive Passes",
    color='white',
    fontsize=14
)

# =====================================
# SHOT MAP (HALF PITCH)
# =====================================

ax3 = fig.add_axes([0.05,0.08,0.42,0.35])

pitch_half = VerticalPitch(
    pitch_type='statsbomb',
    half=True,
    pitch_color='#071A2E',
    line_color='white',
    linewidth=2
)

pitch_half.draw(ax=ax3)

if len(other_shots) > 0:
    pitch_half.scatter(
        other_shots['x'],
        other_shots['y'],
        s=other_shots['shot_statsbomb_xg'] * 3500,
        color='orange',
        edgecolors='white',
        linewidth=1.5,
        ax=ax3
    )

if len(goal_shots) > 0:
    pitch_half.scatter(
        goal_shots['x'],
        goal_shots['y'],
        s=goal_shots['shot_statsbomb_xg'] * 4500,
        marker='*',
        color='gold',
        edgecolors='white',
        linewidth=1.5,
        ax=ax3
    )

ax3.set_title(
    "Shot Map",
    color='white',
    fontsize=14
)

# =====================================
# STATS PANEL
# =====================================

ax4 = fig.add_axes([0.53,0.08,0.42,0.35])

ax4.set_facecolor('#102A43')
ax4.axis('off')

stats_text = f"""
KEY NUMBERS

{ball_receipts} Ball Receipts

{total_passes} Passes
{pass_accuracy}% Accuracy

{progressive_pass_count} Progressive Passes
{progressive_pass_pct}% Progressive

{progressive_carry_count} Progressive Carries
{progressive_carry_pct}% Progressive

{len(shots)} Shots
{len(goal_shots)} Goal


ANALYST TAKEAWAY

Right-sided creator who
combined progression and
goal threat while maintaining
high passing efficiency.
"""

ax4.text(
    0.05,
    0.95,
    stats_text,
    va='top',
    color='white',
    fontsize=15,
    family='monospace'
)

# =====================================
# SAVE
# =====================================

plt.savefig(
    "Lamine_Yamal_Dashboard_Final.png",
    dpi=300,
    bbox_inches='tight',
    facecolor='#071A2E'
)

plt.show()
```
